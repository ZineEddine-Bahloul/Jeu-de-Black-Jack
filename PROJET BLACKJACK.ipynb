## 2.1- A1 - Paquets de cartes

```python
# 1 la fenction PAQUET() :
import random
def paquet() :
    paquet_initial = ""
    paquetCarte = "as 2 3 4 5 9 7 8 9 10 valet dame roi"
    paquetCarte = paquetCarte.split(" ")
    couleursCarte = "pique coeur trefle carreau"
    couleursCarte = couleursCarte.split(" ")
    compteur_paquetCarte = 12
    while compteur_paquetCarte >= 0:
        compteur_couleursCarte = 0
        while compteur_couleursCarte < len(couleursCarte) :
            paquet = paquetCarte[compteur_paquetCarte]+" de "+couleursCarte[compteur_couleursCarte] +","+ paquet_initial
            paquet_initial = paquet
            compteur_couleursCarte = compteur_couleursCarte +1
        compteur_paquetCarte = compteur_paquetCarte -1 
    paquet=paquet[:-1] # pour enlever la derniere vergule
    paquet=list(paquet.split(",")) # trensformer la chaine en liste
    random.shuffle(paquet) # melanger le paquet
    paquet=",".join(paquet) # trensformer la liste en chaine
    return paquet


```

```python
def valeurCarte(carte:str,nomJoueur) :
    assert carte in paquet(),"carte n'existe pas "
    while carte in paquet().split(",") :
        if carte.split(" ")[0] == "as" :
            print(nomJoueur,"Veuillez choisir 1 ou 11 comme valeur de l'as :")
            valeur_du_as = int(input("Valeur du  As?"))
            assert valeur_du_as ==1 or valeur_du_as ==11,"entrez juste 1 ou 11"
            valeurCarte = valeur_du_as
        elif carte.split(" ")[0] in "valet dame roi".split(" ") :
            valeurCarte = 10
        else :
            valeurCarte = int(carte.split(" ")[0])
        break
    return valeurCarte
```

```python
# 3 la fenction initPioche() :
def initPioche(n:int) :
    import random
    initPioche = n * paquet().split(",")
    initPioche = list(initPioche)
    random.shuffle(initPioche)
    return initPioche
```

```python
# 4 la fenction PiocheCarte(p) :
def piocheCarte(p,x=1) :
    compteur_piocheCarte = 0
    liste_vide_piocheCarte= []
    while compteur_piocheCarte <x :
        piocheCarte = [p[compteur_piocheCarte]]+liste_vide_piocheCarte
        liste_vide_piocheCarte = piocheCarte
        p.pop(0)
        compteur_piocheCarte+=1 
    return piocheCarte


    
```

```python
# 1 la fonction initJoueurs :
def initJoueurs(n:int) :
    compteur_joueurs = 1
    liste_initJoueurs = []
    while compteur_joueurs <= n :
        print("entrez le nom du joueur numero",compteur_joueurs,":")
        liste_initJoueurs.append(str(input()))
        compteur_joueurs += 1
    return liste_initJoueurs
```

```python
#2 la fonction initScores :
def initScores(joueurs,v=0) :
    idx = 0
    dictionnaire_initScores = {}
    while idx <len(joueurs) :
        dictionnaire_initScores[joueurs[idx]] = v
        idx += 1
    return dictionnaire_initScores
```

```python
#3 la fenction premier tour 
def premierTour(joueurs,paquetDeCarte) :
    assert type(joueurs)==list,"etrez une liste"
    dico_premierTour = initScores(joueurs)
    cartesDesJoueurs = piocheCarte(paquetDeCarte,2*len(joueurs))
    dico_cartesDesJoueurs = {}
    for cle in dico_premierTour :
        carte1 = piocheCarte(cartesDesJoueurs)[0]
        carte2 = piocheCarte(cartesDesJoueurs)[0]
        dico_cartesDesJoueurs[cle] = carte1+", "+carte2
        dico_premierTour[cle] = valeurCarte(carte1,cle)+valeurCarte(carte2,cle)
    return dico_premierTour,dico_cartesDesJoueurs
```

```python
# 4 la fenction gagnant :

def gagnant(scores) :
    gagnants = []
    score = []
    listeDesNomsDejoueures = list(scores.keys())
    listeDesScorresDejoueures = list(scores.values())
    while max(listeDesScorresDejoueures) > 20 or max(listeDesScorresDejoueures) <= 0 :
        listeDesScorresDejoueures.remove(max(listeDesScorresDejoueures))
        if listeDesScorresDejoueures == [] :
            print("aucun joueur a gagne ")
            break
    compteurListeDesScorresDejoueures = 0
    while compteurListeDesScorresDejoueures < len(listeDesScorresDejoueures) :
        if listeDesScorresDejoueures[compteurListeDesScorresDejoueures]==max(listeDesScorresDejoueures) :
            score.append(max(listeDesScorresDejoueures))
            gagnants.append(listeDesNomsDejoueures[compteurListeDesScorresDejoueures])
        compteurListeDesScorresDejoueures +=1
    dic = {}
    idx = 0
    for i in gagnants:
        dic[i] = score[idx]
        idx+=1
    return dic
gagnant({"y":0,"z":20,"k":0})
```

```python
# une nouvelle version de la fanction  gagnant :
def gagnant21(scores) :
    if type(scores) == dict:
        ret ={None:0}
        gagnants = []
        score = []
        listeDesNomsDejoueures = list(scores.keys())
        listeDesScorresDejoueures = list(scores.values())
        while max(listeDesScorresDejoueures) > 21 or max(listeDesScorresDejoueures) <= 0 :
            listeDesScorresDejoueures.remove(max(listeDesScorresDejoueures))
            if listeDesScorresDejoueures == [] :
                break
        compteurListeDesScorresDejoueures = 0
        while compteurListeDesScorresDejoueures < len(listeDesScorresDejoueures) :
            if listeDesScorresDejoueures[compteurListeDesScorresDejoueures]==max(listeDesScorresDejoueures) :
                score.append(max(listeDesScorresDejoueures))
                gagnants.append(listeDesNomsDejoueures[compteurListeDesScorresDejoueures])
            compteurListeDesScorresDejoueures +=1
        if len(gagnants) != 0 :
            dico = {}
            idx = 0
            for i in gagnants:
                dico[i] = score[idx]
                idx+=1
            ret = dico 
    elif type(scores) == list : 
        score = []
        listeDesScorresDejoueures = list(scores)
        while max(listeDesScorresDejoueures) > 21 or max(listeDesScorresDejoueures) <= 0 :
            listeDesScorresDejoueures.remove(max(listeDesScorresDejoueures))
            if listeDesScorresDejoueures == [] :
                break
        compteurListeDesScorresDejoueures = 0
        while compteurListeDesScorresDejoueures < len(listeDesScorresDejoueures) :
            if listeDesScorresDejoueures[compteurListeDesScorresDejoueures]==max(listeDesScorresDejoueures) :
                score.append(max(listeDesScorresDejoueures))
            compteurListeDesScorresDejoueures +=1
        ret = score
        if ret == [] :
            ret = [min(scores)]
    return ret
```

```python
# une nouvelle version de la fanction  premierTour :
def premierTourInteligent(joueurs,paquetDeCarte) :
    if type(joueurs)!=list :
        joueurs= [joueurs]
    dico_premierTour = initScores(joueurs)
    cartesDesJoueurs = piocheCarte(paquetDeCarte,2*len(joueurs))
    dico_cartesDesJoueurs = {}
    for cle in dico_premierTour :
        carte1 = piocheCarte(cartesDesJoueurs)[0]
        carte2 = piocheCarte(cartesDesJoueurs)[0]
        dico_cartesDesJoueurs[cle] = carte1+", "+carte2
        if valeurCarteInteligent(carte1) == [1,11] and valeurCarteInteligent(carte2) == [1,11] :
            dico_premierTour[cle] = [2,12]
        elif valeurCarteInteligent(carte1) == [1,11] and valeurCarteInteligent(carte2) != [1,11] :
            if valeurCarteInteligent(carte2) == 10 :
                dico_premierTour[cle] = 21
            else :
                dico_premierTour[cle] = [1+valeurCarteInteligent(carte2),11+valeurCarteInteligent(carte2)]
        elif valeurCarteInteligent(carte1) != [1,11] and valeurCarteInteligent(carte2) == [1,11] :
            if valeurCarteInteligent(carte1) == 10 :
                dico_premierTour[cle] = 21
            else :
                dico_premierTour[cle] = [1+valeurCarteInteligent(carte1),11+valeurCarteInteligent(carte1)] 
        else :
            dico_premierTour[cle] = valeurCarteInteligent(carte1)+valeurCarteInteligent(carte2)       
    return dico_premierTour,dico_cartesDesJoueurs
```

```python
def Suite_scores_proba(score_utilisateur,score_max):
    score_max-=1
    proba_mini = 1
    liste_scores = []
    liste_proba = []
    while score_max >= 15 :
        liste_scores.append(score_max)
        liste_proba.append(proba_mini)
        score_max-=1
        proba_mini+=12
    return liste_proba[liste_scores.index(score_utilisateur)]
```

```python
def gagnant_avec_croupir(dico_scores_joueur,dico_Scores_croupier) :
    dico_points_victoires = {}
    listeDesNomsDejoueures = list(dico_scores_joueur.keys())
    listeDesScorresDejoueures = list(dico_scores_joueur.values())
    NomsDuCroupier = list(dico_Scores_croupier.keys())[0]
    ScorresDuCroupier = list(dico_Scores_croupier.values())[0]
    for idx in range(0,len(listeDesScorresDejoueures)) :
        if (listeDesScorresDejoueures[idx] == ScorresDuCroupier) and (ScorresDuCroupier <= 21)  :
            dico_points_victoires[listeDesNomsDejoueures[idx]] = 1
        elif (listeDesScorresDejoueures[idx] < ScorresDuCroupier) and (ScorresDuCroupier <= 21) :
            dico_points_victoires[listeDesNomsDejoueures[idx]] = 0
        elif (listeDesScorresDejoueures[idx] > ScorresDuCroupier) and (listeDesScorresDejoueures[idx]  < 21) :
            dico_points_victoires[listeDesNomsDejoueures[idx]] = 3
        elif (listeDesScorresDejoueures[idx] > ScorresDuCroupier) and (listeDesScorresDejoueures[idx]  == 21)  :
            dico_points_victoires[listeDesNomsDejoueures[idx]] = 5
        elif listeDesScorresDejoueures[idx] < 21  and ScorresDuCroupier > 21 :
            dico_points_victoires[listeDesNomsDejoueures[idx]] = 3
        elif listeDesScorresDejoueures[idx] == 21  and ScorresDuCroupier > 21 :
            dico_points_victoires[listeDesNomsDejoueures[idx]] = 5
        else :
            dico_points_victoires[listeDesNomsDejoueures[idx]] = 0
    return  dico_points_victoires
```

```python
def tourJoueur_avecCroupier(nomJoueur,scoreInitial,jeuDeCartesEnCours,numeroDuTour,dico_cartesJoueurs) :
    ValeurTotalDesCartesPiochesParJoueur = scoreInitial
    if type(scoreInitial) == int :
        ValeurTotalDesCartesPiochesParJoueur = [scoreInitial]
    ValeurTotalDesCartesPiochesParJoueur_copie = max(ValeurTotalDesCartesPiochesParJoueur)
    print("score du joueur",nomJoueur,":", ValeurTotalDesCartesPiochesParJoueur,"t'as les cartes suivantes :",dico_cartesJoueurs[nomJoueur])
    if ValeurTotalDesCartesPiochesParJoueur_copie == 21 :
        ValeurTotalDesCartesPiochesParJoueur = ValeurTotalDesCartesPiochesParJoueur[0]
        print("le joueur",nomJoueur,"est le gagnant")
    elif ValeurTotalDesCartesPiochesParJoueur_copie > 21 :
        ValeurTotalDesCartesPiochesParJoueur = ValeurTotalDesCartesPiochesParJoueur[0]
        print("t'as perdu")
    else :
        while ValeurTotalDesCartesPiochesParJoueur_copie < 21  :
            play = Continue()
            if play==True:
                carte_suivante = piocheCarte(jeuDeCartesEnCours)[0] 
                dico_cartesJoueurs[nomJoueur] = dico_cartesJoueurs[nomJoueur]+", "+carte_suivante
                ValeurTotalDesCartesPiochesParJoueur = definirValeurAs(ValeurTotalDesCartesPiochesParJoueur,carte_suivante)[0]
                if ValeurTotalDesCartesPiochesParJoueur == 21 :
                    print(nomJoueur,"a gagne, t'as eu :",carte_suivante )
                elif ValeurTotalDesCartesPiochesParJoueur > 21 :
                    print("t'as eu",ValeurTotalDesCartesPiochesParJoueur," donc t'as perdu, t'as eu la carte ",carte_suivante)
                else :
                    print("score du joueur",nomJoueur,":", ValeurTotalDesCartesPiochesParJoueur,"t'as les cartes suivantes :",dico_cartesJoueurs[nomJoueur])
                ValeurTotalDesCartesPiochesParJoueur_copie = ValeurTotalDesCartesPiochesParJoueur
            elif play== False :
                ValeurTotalDesCartesPiochesParJoueur =ValeurTotalDesCartesPiochesParJoueur_copie 
                print("t'es sorti du jeu")
                break
    return ValeurTotalDesCartesPiochesParJoueur
```
